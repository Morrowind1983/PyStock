```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
hs300 = pd.read_csv("399300.csv", encoding = "GBK")
cyb = pd.read_csv("399006.csv", encoding = "GBK")
```

```python
hs300["日期"] = pd.to_datetime(hs300["日期"])
hs300.head()
```

```python
cyb["日期"] = pd.to_datetime(hs300["日期"])
cyb.head()
```

```python
hs300_base = hs300["收盘价"][hs300["日期"] == cyb["日期"].min()]
cyb_base = cyb["收盘价"][hs300["日期"] == cyb["日期"].min()]
cyb["收盘价"] = cyb["收盘价"] * hs300_base.iloc[0] / cyb_base.iloc[0]
```

```python
plt.figure(figsize = (18, 6))
plt.plot(hs300["日期"], hs300["收盘价"], label = "沪深300")
plt.plot(cyb["日期"], cyb["收盘价"], label = "创业板")
plt.yscale("log")
plt.legend(loc = 'upper left')
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize = (18, 6))
plt.plot(hs300["日期"], hs300["成交量"])
plt.plot(cyb["日期"], cyb["成交量"])
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize = (18, 6))
plt.plot(hs300["日期"], hs300["成交金额"])
plt.plot(cyb["日期"], cyb["成交金额"])
plt.yscale("log")
plt.show()
```
